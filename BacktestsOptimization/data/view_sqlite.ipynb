```python
import sqlite3

# 连接到您的数据库
conn = sqlite3.connect('binance_futures_data.db')
cursor = conn.cursor()

# 获取所有表名
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("数据库中的表:")
for table in tables:
    print(table[0])

# 对每个表，获取结构和样例数据
for table_name in [t[0] for t in tables]:
    print(f"\n表 '{table_name}' 的结构:")
    cursor.execute(f"PRAGMA table_info({table_name});")
    columns = cursor.fetchall()
    for col in columns:
        print(f"  {col[1]} ({col[2]})")
    
    print(f"\n表 '{table_name}' 的样例数据 (最多5行):")
    cursor.execute(f"SELECT * FROM {table_name} LIMIT 5;")
    rows = cursor.fetchall()
    for row in rows:
        print(f"  {row}")

# 关闭连接
conn.close()
```

```python
# 虽然技术上可以删除它，但不建议这样做
# 删除它可能会导致：
# 1. 查询性能下降
# 2. SQLite可能会在需要时自动重新创建它
# 3. 某些高级索引功能可能会受到影响

# 如果您确实想重置统计信息，更安全的方法是使用ANALYZE命令：
import sqlite3

conn = sqlite3.connect('binance_futures_data.db')
cursor = conn.cursor()

# 重新分析数据库以更新统计信息
cursor.execute("ANALYZE;")
conn.commit()

# 或者，如果您确实要删除它（不推荐）：
# cursor.execute("DROP TABLE sqlite_stat4;")
# conn.commit()

conn.close()
```

```python
import sqlite3

conn = sqlite3.connect('binance_futures_data.db')
cursor = conn.cursor()

# 替换为实际表名
table_name = "YOUR_TABLE_NAME"

# 获取表结构
print(f"表 '{table_name}' 的结构:")
cursor.execute(f"PRAGMA table_info({table_name});")
columns = cursor.fetchall()
for col in columns:
    print(f"  {col[1]} ({col[2]})")

# 获取随机10行样本
print(f"\n表 '{table_name}' 的随机样本数据:")
cursor.execute(f"SELECT * FROM {table_name} ORDER BY RANDOM() LIMIT 10;")
rows = cursor.fetchall()
for row in rows:
    print(f"  {row}")

conn.close()
```
